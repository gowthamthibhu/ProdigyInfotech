```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
train = pd.read_csv(r"C:\Users\gowth\OneDrive\Desktop\ProdigyInfotech\PRODIGY_ML_01\dataset\train.csv")
test = pd.read_csv(r"C:\Users\gowth\OneDrive\Desktop\ProdigyInfotech\PRODIGY_ML_01\dataset\test.csv")
train.head()
```

```python
x = train[['LotArea', 'BedroomAbvGr', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath']]
Y = train['SalePrice']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, Y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
mse = mean_squared_error(ytest, ypred)
print("Mean Squared Error = ", mse)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(ytest, ypred)
plt.plot([ytest.min(), ytest.max()], [ytest.min(), ytest.max()], 'r--', lw=2)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.grid(True)
plt.show()

for train in ['LotArea', 'BedroomAbvGr', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath']:
    plt.figure(figsize=(10, 6))
    plt.scatter(xtest[train], ypred)
    plt.xlabel(train)
    plt.ylabel('Predicted Prices')
    plt.title(f'{train} vs Predicted Prices')
    plt.grid(True)
    plt.show()  
```
