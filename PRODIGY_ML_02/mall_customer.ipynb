```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\gowth\OneDrive\Desktop\ProdigyInfotech\PRODIGY_ML_02\dataset\Mall_Customers.csv"
data = pd.read_csv(file_path)
```

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids',marker='x')
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=data, s=100)
plt.title('Customer Segments')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
